```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
d = pd.read_csv(r"C:\Users\Lenovo\Desktop\gaurav\tips.csv")
d.columns
```

```python
sns.countplot(x = d["sex"],color="g")  # its create graphy in sns
sns.countplot(x = d["day"],color="y")
sns.countplot(x = d["smoker"] ,color="r")
```

```python
""" 
value_counts() =>
                value_counts its count numbers of element
                """
d['smoker'].value_counts()
d["sex"].value_counts()        # its return how many male and how female present in column 
```

```python
""" Plot =>
             its also create graph form data  its attributes like kind,bar,barh
             bar plot provide attributes like color,edgecolor,width,alpha,grid,xlabel,ylabel,grid """
#d["size"].plot(kind = "barh")
d['day'].value_counts().plot(kind = 'bar',color="yellow",edgecolor ="black" ,width =0.2,alpha=0.4,xlabel =" its day ",ylabel="size of day",grid =True) 
colors = ["red","black","skyblue","purple"]
d['day'].value_counts().plot(kind = 'bar',color = colors ) #use multiple color in one graph
```

```python
""" pie =>
           pies chart also provide more attributes like kind,title,explode,wedgeprops wedgeprops its use to give border on 
           circle and explode is use to show cut sapce """
#make a pie chart
#d['day'].value_counts().plot(kind = 'pie' , autopct = '%.2f') 
d["day"].value_counts().plot(kind = "pie", autopct ="%.2f%%" ,explode =[0.05,0.01,0.04,0.03] ,wedgeprops={'edgecolor':"black",'linewidth':1.6})
```

```python
""" distplot =>
           displot in seaborn its also use for made graph and its provide attributes
           like hist,kde,vertical,rug,hist_kws,kde_kws rug is use for show small markes individual point, """
# Making a dist plot
sns.distplot(d['tip'],
            hist = True,
            kde = True,
            rug = True,     # use for show small markes individual point
            vertical = False,     #its use for vertical grpah
            hist_kws = {'edgecolor':'black'},     # its use for border color
            kde_kws ={'linewidth':1.4})            #linewidth size
#d['tip']
```

```python
"""" 3 . Boxplot =>  
                    boxplot its provide attributes like color ,notch,,flierprops,boxprops,
                    palette and others 
for find our outliers
1 . lower fence
2 . 25% data
3. IQR(Inter Quartile Range)(75%-25%)
4. 75% data
5. upper fence """
sns.boxplot(d['size'])
sns.boxplot(d["size"],color ="lightgreen",
           notch =True,                               #add notch around median
            flierprops={'marker':"o",'color':"r"},      # customize outliers
            boxprops={'facecolor':'yellow','edgecolor':'black'},  #its customizes box(border,linewidth)
            palette="coolwarm")       #palette its set color for multiple box
```

```python
"""" Scatterplot =>
                its also work with two column and its provide attributes like color,size
                ,sizes,palette,linewidth,alpha others 
                hue,data,style

               palette :=> dark,coolwarm,bright,muted,pastel,deep
    scatterplot(Numerical column - Numerical column) 
"""
#sns.scatterplot(x = d['total_bill'] , 
  #              y = d['tip'])
sns.scatterplot(data = d,x=d["total_bill"],y=d["tip"],hue =d["sex"],
                style = d['smoker'],
                size="total_bill",       #vary point size by anoter variable
                sizes=(23,344),           # range of marker sizes
                palette ="deep")     # its use to control set of colors used for categories defined by hue parameter
```

```python
"""
Barplot =>
           barplot its also provide many attributes like color,palette,orient,order,hue
           ,hue_order and others 
Barplot (Numerical - Categorical) 
"""
sns.barplot(x = d['sex'] , y = d['time'],
            palette ="dark",
            orient ="h",        # its use for h,v
            hue_order =["Female","Male"])     # its use for give order in hue
```

```python
"""
heat Plot =>
               its also import in plot and its provide many attributes like 
               annot,cbar,cmap,fmt,linewidth and others 
"""
p = pd.crosstab(d['smoker'] , d['day'])
p
sns.heatmap(p,annot =True,          #annot its use for number in area
            fmt ='d',                 # its use for format
            cbar = False,             # its use for hide line 
            cmap ="coolwarm")     # cmap its use for color like palette
```
