```python
""" isnull =>
         its use to find null value in data and record """

""" notnull =>
               its use also found not null values """
a= {
    "Rank":[1,None,3,4,5,6,7,8],
    "Actor":['Shah Rukh Khan','Salman Khan','Ajay Devgn','Akshay Kumar','Sanjay Dutt','Aamir Khan','Hritik Roshan','Ranbir Kapoor'],
    "genre":['romance','action',None,'action/drama','drama','action','action/drama','romance'],
    "hits":[45,56,None,45,78,None,57,89]
}
v   = pd.DataFrame(a)
v.isnull()     # its return true and false if its found None(null value) its return false
v.isnull().sum()       # when use sum with null values its return number of null values

v.notnull()         # its also familer like isnull
```

```python
""" dropna =>
              its use to delete null values from the record """
a= {
    "Rank":[1,None,3,4,5,6,7,8],
    "Actor":['Shah Rukh Khan','Salman Khan','Ajay Devgn','Akshay Kumar','Sanjay Dutt','Aamir Khan','Hritik Roshan','Ranbir Kapoor'],
    "genre":['romance','action',None,'action/drama','drama','action','action/drama','romance'],
    "hits":[45,56,None,45,78,None,57,89]
}
v   = pd.DataFrame(a)
v.dropna()    # its delete null values
v.dropna(axis=1)  #its delete all column who take it null values in data
```

```python
""" dropna =>
              dropna with axis, thresh use in datasets
             axis = 0 is hide all null values column
             axis = 1 is also hide but not all
             thresh = its use to hide column according thier null values
"""
d = pd.read_csv(r"C:\Users\Lenovo\Desktop\gaurav\titanic.csv")
d.dropna(axis= 0)
d.dropna(axis= 1,thresh =2)
```

```python
""" fillna => 
              fillna its use to fill some value in data in place of null values """
a= {
    "Rank":[1,None,3,4,5,6,7,8],
    "Actor":['Shah Rukh Khan','Salman Khan','Ajay Devgn','Akshay Kumar','Sanjay Dutt','Aamir Khan','Hritik Roshan','Ranbir Kapoor'],
    "genre":['romance','action',None,'action/drama','drama','action','action/drama','romance'],
    "hits":[45,56,None,45,78,None,57,89]
}
v   = pd.DataFrame(a)
v.fillna(0)  # its fill 0 in place of null values
```
